###str.extract(정규표현식)<br>
###정규 표현식에 부합되는 문자열을 추출.

```python
import pandas as pd
from sklearn.datasets import fetch_20newsgroups
```

```python
news=fetch_20newsgroups(subset='train')
```

```python
df = pd.DataFrame(news.data, columns=['data'])
```

```python
# 이메일 주소 추출하는 정규표현식
E = '([a-zA-Z0-9+._]+@[a-zA-Z0-9._-]+\.[a-zA-Z0-9_-]+)'
```

```python
# str.extract
df['email'] = df['data'].str.extract(E)
```

```python
# 이메일 주소없는 행 제거
df.dropna(subset=['email'], inplace=True)
```
